# Pseudo Facebook DataSet

### Variables Description 

1. <b>userid</b>: ID of the user
2. <b>age</b>: User's Age(years)
3. <b>dob_day</b>: Day of date of birth(1-31)
4. <b>dob_year</b>: Year of date of birth
5. <b>dob_month</b>: Month of date of birth
6. <b>gender</b>: M/F
7. <b>tenure</b>: How long have facebook user been on site
8. <b>friend_count</b>: Total number of friends
9. <b>friendships_initiated</b>: Friend Requests Send
10. <b>likes</b>: Total number of likes by User
11. <b>likes_received</b>: Total number of likes received by User
12. <b>mobile_likes</b>: No of likes by user(through mobile)
13. <b>mobile_likes_received</b>: No of likes received by user(through mobile)
14. <b>www_likes</b>: No of likes by user(through desktop)
15. <b>www_likes_received</b>: No of likes received by user(through desktop)

### Tenure

```python
facebook_df[facebook_df.tenure.isnull()]
```

- Both missing values for females
- Age group is above 65 years

- If the missing value count would have been large <b>KNN imputation</b> works best for treating missing values

```python
#Manual Imputation
tenure1=facebook_df.tenure[(facebook_df.age==88) & (facebook_df.gender=='female')].mean()
facebook_df.tenure[facebook_df.userid==1414063].fillna(tenure1)

tenure2=facebook_df.tenure[(facebook_df.age==103) & (facebook_df.gender=='female')].mean()
facebook_df.tenure[facebook_df.userid==2095829].fillna(tenure2)
```
